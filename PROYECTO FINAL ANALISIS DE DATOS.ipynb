<p>El dióxido de carbono (fórmula química CO2) es un compuesto de carbono y oxígeno que existe como gas incoloro en condiciones de temperatura y presión estándar (TPS). Está íntimamente relacionado con el efecto invernadero. La concentración actual es de alrededor 0,04 % (410 ppm) en volumen, un 45 % mayor a los niveles preindustriales de 280 ppm. Fuentes naturales incluyen volcanes, aguas termales, géiseres y es liberado por rocas carbonatadas al diluirse en agua y ácidos. Dado que el CO2 es soluble en agua, ocurre naturalmente en aguas subterráneas, ríos, lagos, campos de hielo, glaciares y mares. Está presente en yacimientos de petróleo y gas natural.3</p>

```python
import matplotlib.cbook as cbook
 
fig, ax = plt.subplots()
ax.scatter(df_emisiones1['Año'], df_emisiones1['Total'], alpha=1)
ax.set_xlabel("Año", fontsize=15)
ax.set_ylabel("Emisiones de Dioxido de carbon", fontsize=15)
ax.set_title('Revisión de emisiones de Dioxido de Carbon por año')
 
ax.grid(True)
fig.tight_layout()
 
plt.show()
```

```python
#Se puede ver que entre el año 2010 y el 2015 se da la mayor alza de emisones de dioxido de carbon.
```

```python
import matplotlib.cbook as cbook
 
fig, ax = plt.subplots()
ax.scatter(df_emisiones1['Año'], df_emisiones1['Sub total industrias de energía'], alpha=1)
ax.set_xlabel("Año", fontsize=15)
ax.set_ylabel("Emisiones de Dioxido de carbon por sector industrial", fontsize=15)
ax.set_title('Revisión de emisiones de Dioxido de Carbon por año del sector industrial')
 
ax.grid(True)
fig.tight_layout()
 
plt.show()
```

```python
import matplotlib.cbook as cbook
 
fig, ax = plt.subplots()
ax.scatter(df_emisiones1['Año'], df_emisiones1['Sub total sector consumo'], alpha=1)
ax.set_xlabel("Año", fontsize=15)
ax.set_ylabel("Emisiones de Dioxido de carbon por sector consumo", fontsize=15)
ax.set_title('Revisión de emisiones de Dioxido de Carbon por año del sector consumo')
 
ax.grid(True)
fig.tight_layout()
 
plt.show()
```
